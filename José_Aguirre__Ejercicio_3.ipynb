### Ejercicio 1.5

Se le llama "Camel Case" a la convención de escribir una frase sin espacios, pero marcando el inicio de cada palabra poniendo su primera letra en mayúscula. Por ejemplo, la frase
```
"  Algoritmos y    estructuras de   datos   "
```
transformada a Camel Case queda así:
```
"AlgoritmosYEstructurasDeDatos"
```

Escriba una función que transforme a Camel Case y pruébela:

````python
def CamelCase(s):
    """Retorna un string conteniendo la versión Camel Case del string s"""
    frase = ''
    estado = 'FUERA' # tiene que empezar en fuera pues de esta manera en el primer elemento puede entrar si es que el str comienza con una letra
    for k in range(0,len(s)):
      if estado == 'DENTRO':
        if s[k] != ' ':
          frase = frase + s[k] # agregamos la letra a la frase
        else:
          estado = 'FUERA'
      else: #estado == 'FUERA'
        if s[k] != ' ':
          estado = 'DENTRO'
          frase = frase + s[k].capitalize() #Aca pasamos de un ' ' a una letra, entonces es el inicio de una palabra, por ende colocamos esta primera palabra en mayuscula.
    return frase
````

````python
print(CamelCase("    Algoritmos y    estructuras de   datos   "))
````
